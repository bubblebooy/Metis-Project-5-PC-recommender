## Combine

```python
def storage_combined_details(storage_df):
    storage_df = storage_df.copy()
#     storage_df['price'] = storage_df['price'].where( storage_df['price'].str.contains('$',regex=False) , None)
    storage_df['storage_price'] = pd.to_numeric(storage_df['price'].str.slice(start=1),errors='coerce')
    storage_df['storage_price'] = storage_df['storage_price'].where(storage_df['storage_price'] != 0, np.nan).fillna(storage_df['storage_price'].mean())
    
    _df = storage_df['name'].str.replace(r'(\d*) TB',r'\1kkk TB').str.replace(r'kkk TB',r'000 GB')
    _df = pd.to_numeric( _df.str.extract(r'(\d* GB)',expand=False).str.slice(stop=-3) )
    storage_df['GB'] = _df
    storage_df['SSD'] = storage_df['name'].str.contains('SSD')
    storage_df['NVME_M2'] = storage_df['name'].str.contains('NVME') | storage_df['name'].str.contains('M.2')
    
    group = storage_df.groupby('build_url')
    storage_combined_df = group[['storage_price','GB']].sum()
    storage_combined_df[['SSD','NVME_M2']] = group[['SSD','NVME_M2']].any()
    storage_combined_df['storage_count'] = group['storage_price'].count()
    return storage_combined_df

storage_combined_details(df_part_build[df_part_build['type'] == 'Storage']).head(3)
```

```python
df_combined =  pd.concat([
    df_builds[['build_url','name','price']].set_index('build_url'),
    df_builds_details[['build_url','description','Date Published','CPU Clock Rate','GPU Effective Memory Clock Rate','GPU Core Clock Rate']].set_index('build_url'),
    ],
    axis=1,join='inner')

df_combined =  pd.merge( df_combined,
                         df_part_build[df_part_build['type'] == 'CPU'][['build_url','name','part_url','price']],
                         how='inner',left_index=True,right_on='build_url',suffixes=('_build','_cpu'))
df_combined =  pd.merge( df_combined,
                         df_cpu_details.drop(columns=['Model','Maximum Supported Memory','Part #']),
                         how='inner',on='part_url')

df_combined =  pd.merge( df_combined,
                        storage_combined_details(df_part_build[df_part_build['type'] == 'Storage']),
                        how='inner', left_on='build_url', right_index=True)
                        

part_types = ['Motherboard', 'Memory', 'Video Card','Case','Power Supply', 'CPU Cooler', 'Fan Controller','Sound Card', 'Wireless Network Adapter', 'Laptop','Optical Drive', 'Wired Network Adapter']
for part_type in part_types:
    _df = df_part_build[df_part_build['type'] == part_type].groupby('build_url').first()
    _df = _df[['name']].rename(columns={"name": part_type}).reset_index()
    df_combined =  pd.merge( df_combined,
                         _df,
                         how='left',on='build_url')
    
part_types = ['Memory', 'Video Card']
for part_type in part_types:
    _df = df_part_build[df_part_build['type'] == part_type].groupby('build_url').count()
    _df = _df[['name']].rename(columns={"name": part_type + "_count"}).reset_index()
    df_combined =  pd.merge( df_combined,
                         _df,
                         how='left',on='build_url')
    
```

```python
def str_GHz(series):
    mask_MHz = series.str.contains('MHz')
    mask_THz = series.str.contains('THz')
    series = pd.to_numeric( series.str.slice(stop=-4))
    series = (series/1000).where(mask_MHz, series)
    series = (series*1000).where(mask_THz, series)
    return series
```

```python
def cache_to_num(series):
    split = series.str.split(' x ',expand=True)
    split[0] = pd.to_numeric(split[0], errors = 'coerce')
    split[1] = pd.to_numeric( split[1].str.replace(' MB','000').str.replace(' kB',''), errors = 'coerce')
    return split[0] * split[1]
# cache_to_num(df_combined['L3 Cache'])
```

# Recommender

```python
class Recommender:
    def __init__(self, drop_columns = [], distance_metric = pairwise.euclidean_distances, preprocessors = [StandardScaler()],feature_weights = {}):
        self.distance_metric = distance_metric
        self.build_list = None
        self.drop_columns = drop_columns
        self.preprocessors = preprocessors
        if not self.preprocessors:
            self.preprocessors = []
        self.feature_weights = feature_weights
        
        
    def check_is_fitted(self):
        if self.build_list is None:
            raise Exception("""This Recommender instance is not fitted
            yet. Call 'fit' with a build_list as an argument before using this
            estimator.""")
        
    def set_drop_columns(self, drop_columns):
        self.drop_columns = drop_columns
        
    def preprocesses(self, build_list):
        build_list = build_list.drop(columns = self.drop_columns, errors='ignore').dropna()
        build_list_cols = build_list.columns
        for preprocessor in self.preprocessors:
            build_list = preprocessor.transform(build_list)
        for feature in self.feature_weights:
            try:
                build_list = pd.DataFrame(build_list, columns = build_list_cols)
                build_list[feature] = build_list[feature] * self.feature_weights[feature]
            except:
                raise Exception("feature_weights do not work with preprocessors the change the number of features")
            
        return build_list
        #transform standardization 
        #transform dim reduction?
    
    def fit(self, build_list):
        self.build_list = build_list
        
        build_list = build_list.drop(columns = self.drop_columns, errors='ignore').dropna()
        
        for preprocessor in self.preprocessors:            
            build_list = preprocessor.fit_transform(build_list)
#         if self.scaler:
#             self.scaler.fit(build_list.drop(columns = self.drop_columns, errors='ignore').dropna())
        #fit standardization 
        #fit dim reduction?
        
    def recommend(self, input_builds, fixed_columns = {}, return_type = 'index'):
        self.check_is_fitted()
        if isinstance(input_builds , pd.Series):
            input_builds = input_builds.to_frame().T
        elif not isinstance(input_builds , pd.DataFrame):
            raise Exception('input_builds should be a pd.Series or pd.DataFrame')
        build_list = self.build_list
        for col in fixed_columns:
            build_list = build_list[build_list[col] == fixed_columns[col]]
        
        distances = self.distance_metric(X = self.preprocesses(build_list),
                                         Y = self.preprocesses(input_builds))
        distances = distances.sum(axis = 1)
        
        if return_type == 'index':
            return self.build_list.index[distances.argsort()]
        elif return_type == 'distances':
            return distances

        return distances.argsort()
```

## NN encoder / decoder

```python
print(df_X.drop(columns = ['Date Published','price_build','number_ratings','avg_rating','storage_price']).shape)
reg_cols = ['GPU Effective Memory Clock Rate','GPU Core Clock Rate','Core Count',
                 'Core Clock','Boost Clock','TDP','L2 Cache','L3 Cache','Lithography',
                 'GB','storage_count','Memory','Power Supply','Video Card_count','Video Card_memory']
df_y_reg = df_X[reg_cols]
scaler_reg = RobustScaler()
df_y_reg = scaler_reg.fit_transform(df_y_reg)

log_cols = ['Simultaneous Multithreading','Manufacturer','SSD','NVME_M2','Motherboard','CPU Cooler','Fan Controller','Sound Card',
            'Wireless Network Adapter','Laptop','Optical Drive','Wired Network Adapter','Video Card_AMD']
df_y_log = df_X[log_cols]

series_cols = ['Series_AMD Ryzen 3', 'Series_AMD Ryzen 5',
               'Series_AMD Ryzen 7', 'Series_AMD Ryzen 9', 'Series_Intel Core i3',
               'Series_Intel Core i5', 'Series_Intel Core i7', 'Series_Intel Core i9',
               'Series_other']
df_y_series = df_X[series_cols]

motherboard_cols = ['Motherboard_Manufacturer_ASRock','Motherboard_Manufacturer_Asus', 'Motherboard_Manufacturer_Gigabyte',
                    'Motherboard_Manufacturer_MSI', 'Motherboard_Manufacturer_other']
df_y_motherboard = df_X[motherboard_cols]

print(df_y_reg.shape)
print(df_y_log.shape)
print(df_y_series.shape)
print(df_y_motherboard.shape)
```

```python
def create_autoencoder(weights = None):
    inputs = Input(shape=(43,))
    # x = Dropout(0)(inputs)
    x = Dense(30 , activation = "relu")(inputs)
    x = Dense(25 , activation = "relu")(x)
    x = Dense(20 , activation = "relu")(x)
    x = Dense(25 , activation = "relu")(x)
    x = Dense(30 , activation = "relu")(x)
    x = Dense(60 , activation = "relu")(x)
    x = Dense(120 , activation = "relu")(x)

    out_reg = Dense(15 , activation = "linear", name = 'out_reg')(x)
    out_log = Dense(13 , activation = "sigmoid", name = 'out_log')(x)
    out_series = Dense(9 , activation = "softmax", name = 'out_series')(x)
    out_motherboard = Dense(5 , activation = "softmax", name = 'out_motherboard')(x)
    model = Model(inputs=inputs, outputs=[out_reg, out_log, out_series, out_motherboard]) #
    model.compile(optimizer=Adam(), #learning_rate=0.0001
                      loss=['mean_squared_error','binary_crossentropy','categorical_crossentropy','categorical_crossentropy']) #
    if weights is not None:
        model.load_weights(weights)
    return model
model = create_autoencoder()
```

```python
X = df_X.drop(columns = ['Date Published','number_ratings','avg_rating','storage_price'])
X['price_build'] = X['price_build'].fillna(X['price_build'].mean())
X[reg_cols] = df_y_reg

history = model.fit(X, y = [df_y_reg,df_y_log,df_y_series,df_y_motherboard],epochs=50,validation_split=0.15) #
model.save_weights('autoencoder.h5')
```

```python
fig=plt.figure(figsize = (18,12))
i = 1
for loss in history.history:
    if not 'val' in loss:
        ax = plt.subplot(3,3,i)
        ax.plot(history.history[loss])
        ax.plot(history.history['val_' + loss])
        plt.ylim(0,min(50,plt.ylim()[1]))
        i += 1
```
